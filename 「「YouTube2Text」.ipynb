```python
from yt_dlp import YoutubeDL
import ffmpeg
from tqdm import tqdm

def download_audio(url, output_dir='audio'):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': f'{output_dir}/%(id)s.%(ext)s',
        'ffmpeg_location': '/usr/bin/ffmpeg',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
    }

    with tqdm(total=1, desc='Downloading') as pbar:
        with YoutubeDL(ydl_opts) as ydl:
            try:
                ydl.download([url])
                pbar.update(1)
                print('Download complete!')
            except Exception as e:
                print(f"An error occurred: {e}")

# 使用範例
url = 'https://youtu.be/Zc9FJo9rd_g?si=ZO-XJ0DSKVb2pxiB'
download_audio(url)
```

```python
from pydub import AudioSegment
from transformers import pipeline
import os

# 加載 Hugging Face 的語音轉錄模型
transcriber = pipeline("automatic-speech-recognition", model="openai/whisper-large-v3-turbo")

```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)


def process_audio(audio_path, segment_length, output_dir='segments'):
    # 確保輸出目錄存在
    os.makedirs(output_dir, exist_ok=True)
    transcript_all = ""

    # 加載音頻
    audio = AudioSegment.from_file(audio_path)

    # 將音頻分割成段落
    for i in range(0, len(audio), segment_length):
        chunk = audio[i:i + segment_length]
        chunk_path = os.path.join(output_dir, f"output_{i // segment_length}.wav")
        chunk.export(chunk_path, format="wav")  # Hugging Face 模型支持 WAV 格式

        # 使用 Hugging Face 模型轉錄
        try:
            result = transcriber(chunk_path)  # 無需設置 language 參數
            transcript = result['text']  # 獲取轉錄文本
            transcript_all += f" {transcript}"
        except Exception as e:
            print(f"段落 {i // segment_length} 處理時出現錯誤: {e}")

    return transcript_all

# 使用範例
audio_path = "audio/Zc9FJo9rd_g.mp3"  # 音頻文件
segment_length = 30000  # 分段長度（例如 30 秒）
final_transcript = process_audio(audio_path, segment_length)
print("完整轉錄結果：", final_transcript)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.text_splitter import RecursiveCharacterTextSplitter  # 添加此行
import gradio as gr

# 生成嵌入並儲存
def generate_embeddings_and_store(texts, api_key):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", google_api_key=api_key)
    vector_index = Chroma.from_texts(texts, embeddings).as_retriever(search_kwargs={"k": 5})
    return vector_index

# 問題回答
def get_answer(question, vector_index, api_key):
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", api_key=api_key)
    qa_chain = RetrievalQA.from_chain_type(
        llm,
        retriever=vector_index,
        return_source_documents=True
    )
    result = qa_chain({"query": question})
    return result["result"]

# 文本處理及分割
def process_and_split_text(transcript, chunk_size=1000, chunk_overlap=200):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    texts = text_splitter.split_text(transcript)
    return texts

# 使用範例
final_transcript = final_transcript  # 假設這是您的轉錄結果

# 1. 生成並儲存嵌入
api_key = 'AIzaSyA_7Cgqi7M_ZVD_tVQKNMISksnE7Cd9NSQ'  # 替換成你的 Google API 密鑰
texts = process_and_split_text(final_transcript)  # 將文本分割為小塊
vector_index = generate_embeddings_and_store(texts, api_key)

# 2. Gradio 介面
with gr.Blocks() as demo:
    gr.Markdown("##  Q&A")
    question_input = gr.Textbox(label="Enter your question about the transcript:")
    answer_output = gr.Textbox(label="Answer")

    def query(question):
        return get_answer(question, vector_index, api_key)

    question_input.submit(query, inputs=[question_input], outputs=[answer_output])

demo.launch()
```
